#Feature Scaling


Feature Scaling is the last step we do in a Feature Engineering Pipeline. It is one of the easiest steps in Feature Engineering.

Feature Scaling - Feature Scaling is a Technique to standardize the independent features present in the data in a fixed range.


Types of Feature Scaling : 1.) Standardization
                           2.) Normalization

##Feature Scaling - Standardization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Social_Network_Ads.csv")
```

```python
df.sample(5)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Purchased', axis=1), df['Purchased'], test_size=0.3, random_state=0)
```

```python
X_train.shape, X_test.shape
```

Standard Scaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
X_train_scaled
```

```python
np.round(X_train.describe(), 1)
```

```python
np.round(X_train_scaled.describe(), 1)
```

Effect Of Scaling

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(x=X_train['Age'], y=X_train['EstimatedSalary'], color='green')
ax1.set_title('Before Scaling')
ax2.scatter(x=X_train_scaled['Age'], y=X_train_scaled['EstimatedSalary'], color='red')
ax2.set_title('After Scaling')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize=(12,5))

ax1.set_title('Before Scaling')
sns.kdeplot(data=X_train['Age'], ax=ax1)
sns.kdeplot(data=X_train['EstimatedSalary'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(data=X_train_scaled['Age'], ax=ax2)
sns.kdeplot(data=X_train_scaled['EstimatedSalary'], ax=ax2)
plt.show()
```

Comparison of Distributions

Why Scaling Is Important?

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
lr_scaled.fit(X_train_scaled, y_train)
```

```python
y_pred = lr.predict(X_test)
y_pred_scaled = lr_scaled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Actual :", accuracy_score(y_test, y_pred))
print("Scaled :", accuracy_score(y_test, y_pred_scaled))
```

##Feature Scaling - Normalization

Normalization is a Technique often applied as part of data preparation for machine learning. The goal of normalization is to change the values of numeric columns in the dataset to use a common scale, without distorting differences in the ranges of values or losing information.

- MinMax Scaling. (Most Important)
- Mean Normalization.
- Max AbsScaling.
- Robust Scaling.
